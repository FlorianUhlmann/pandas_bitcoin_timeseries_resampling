## Import the BTCUSD data

```python
import pandas as pd

df_M1 = pd.read_csv('gemini_BTCUSD_2019_1min.csv',header=1) 
df_M1.tail(11)
```

### set "Date" column as pandas.DatetimeIndex

```python
#set column 'time' from type array to datetime
df_M1['Date'] = pd.to_datetime(df_M1['Date'])
#set column 'time' to index*
df_M1 = df_M1.set_index('Date')
df_M1.tail(10)
```

### change timeframe

```python
#drop and sort 
dataframe_M1  = df_M1.resample('1Min').agg({'Open': 'first', 'High': 'max', 'Low': 'min', 'Close': 'last','Volume':'sum'})
dataframe_M5  = df_M1.resample('5Min').agg({'Open': 'first', 'High': 'max', 'Low': 'min', 'Close': 'last','Volume':'sum'})
dataframe_M5_w_symbol  = df_M1.resample('5Min').agg({'Open': 'first', 'High': 'max', 'Low': 'min', 'Close': 'last','Volume':'sum','Symbol':'last'})
dataframe_M15 = df_M1.resample('15Min').agg({'Open': 'first', 'High': 'max', 'Low': 'min','Close': 'last','Volume':'sum'})
dataframe_M30 = df_M1.resample('30Min').agg({'Open': 'first', 'High': 'max', 'Low': 'min', 'Close': 'last','Volume':'sum'})
dataframe_H1  = df_M1.resample('60Min').agg({'Open': 'first', 'High': 'max', 'Low': 'min', 'Close': 'last','Volume':'sum'})
dataframe_H4  = df_M1.resample('240Min').agg({'Open': 'first', 'High': 'max', 'Low': 'min', 'Close': 'last','Volume':'sum'})
dataframe_1D  = df_M1.resample('1D').agg({'Open': 'first', 'High': 'max', 'Low': 'min', 'Close': 'last','Volume':'sum'})
dataframe_1W  = df_M1.resample('1W').agg({'Open': 'first', 'High': 'max', 'Low': 'min', 'Close': 'last','Volume':'sum'})
dataframe_1B = df_M1.resample('1B').agg({'Open': 'first', 'High': 'max', 'Low': 'min', 'Close': 'last','Volume':'sum'})
```
